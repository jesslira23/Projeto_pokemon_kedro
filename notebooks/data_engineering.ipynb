```python
def calculate_number_of_abilities(df):
    abilities = ast.literal_eval(df['abilities'])
    number_of_abilities = len(abilities)
    return number_of_abilities
```

```python
pp_pokemon.assign(
    number_of_abilities = pp_pokemon.apply(
        calculate_number_of_abilities,
        axis=1
    )
)
```
