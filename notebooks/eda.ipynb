```python
dados_pokemon_graph = df[['type1','weight_kg']].groupby('type1').mean().sort_values('weight_kg',ascending=False).reset_index()
```

```python
ax = plt.subplots(figsize=(16, 8))
sns.barplot(x='type1',y='weight_kg',data=dados_pokemon_graph)
```

```python
ax = plt.subplots(figsize=(16, 8))
dados_pokemon_graph = df[['type1','attack']].groupby('type1').mean().sort_values('attack',ascending=False).reset_index()
sns.barplot(x='type1',y='attack',data=dados_pokemon_graph)
```

```python
corr = df[['weight_kg','attack']].corr()   #correlacao

ax = plt.subplots(figsize=(16, 8))

sns.heatmap(data = corr
            , annot=True
            , annot_kws={'size': 24}

)
```
